## Lists
- List items are ordered, changeable, and allow duplicate values.  
- List items are indexed, the first item has index [0], the second item has index [1] etc.
- List can be of any data type and can contain multiple data types

```python
# Lists

myList = ["apple", "banana", "cherry", "apple", "cherry", "banana", "orange", "orange"]
print(myList)
print(len(myList))
print(type(myList))

anotherList = list(("apple", "banana", "cherry"))
print(anotherList)
```

```python
# Accessing list items
myList = ["apple", "banana", "cherry", "apple", "cherry", "banana", "orange"]

print(myList[1])
print(myList[-1])
print(myList[1:4])
print(myList[:2])
print(myList[4:])
print(myList[-3:-1])

if "apple" in myList:
    print("Yes")
else:
    print("No")
```

```python
# Adding and inserting to lists

myList = ["apple", "banana", "cherry", "orange"]
myList[1] = "blueberry"
print(myList)

myList[1:3] = "mango", "kiwi"
print(myList)

# What happens to this list?
myList[1:3] = ["watermelon"]
print(myList)

myList.insert(2, "kiwi")
print(myList)

myList.append("kiwi")
print(myList)

tropical = ["mango", "pineapple", "papaya"]
myList.extend(tropical)
print(myList)

# Can I do this?
myList = ["apple", "banana", "cherry", "orange"]
tropical = ("mango", "pineapple", "papaya")
myList.extend(tropical)
print(myList)
```

```python
# Removing from a list

myList = ["apple", "banana", "cherry", "orange", "banana"]
myList.remove("banana")
print(myList)

myList.pop(1)
print(myList)

myList.pop()
print(myList)

del myList[1]
print(myList)

myList.clear()
print(myList)
```

```python
# Looping through lists

myList = ["apple", "banana", "cherry", "orange", "banana"]
for x in myList:
    print(x)

print('')

for i in range(len(myList)):
    print(myList[i])

print('')

i = 0
while i < len(myList):
    print(myList[i])
    i = i + 1
```

```python
# List Comprehensions
# List comprehension offers a shorter syntax when you want to create a new list 
# based on the values of an existing list.
# Example:
# Based on a list of fruits, you want a new list, 
# containing only the fruits with the letter "a" in the name.
# Without list comprehension you will have to write a for statement with a conditional test inside:

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)
print(newlist)

# With List comprehension

fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)

# The Syntax
# newlist = [expression for item in iterable if condition == True]
# Similar to Java lambdas

# Condition
# The condition is like a filter that only accepts the items that valuate to True.

newlist = [x for x in fruits if x != 'apple']
print(newlist)

numList = [x for x in range(10)]
print(numList)

numList = [x for x in range(10) if x > 5]
print(numList)

newList = [x.upper() for x in fruits]
print(newList)

newList = ['empty' for x in fruits]
print(newList)

# if not banana then return x else return orange
newList = [x if x != "banana" else "orange" for x in fruits]
print(newList)

# original list still remains the same!
print(fruits)

numbers = range(1, 11)
even = [x for x in numbers if x % 2 == 0]
print(even)

numbers = range(1, 11)
nums = ["even" if x % 2 == 0 else "odd" for x in numbers]
print(nums)

```

```python
# Sorting lists

fruits = ["banana", "mango", "apple", "cherry", "kiwi"]
fruits.sort()
print(fruits)

numList = [100, 50, 65, 82, 23]
numList.sort()
print(numList)

numList = [100, 50, 65, 82, 23]
numList.sort(reverse=True)
print(numList)

def myFun(n):
    return abs(n - 50)

numList = [100, 50, 65, 82, 23]
numList.sort(key=myFun)
print(numList)

# careful of Case when sorting
fruits = ["banana", "Mango", "apple", "cherry", "Kiwi"]
#fruits.sort()
fruits.sort(key = str.lower)
print(fruits)
```

```python
# Joining lists
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
list3 = list1 + list2
print(list3)

list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
for x in list1:
    list2.append(x)
print(list2)

list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
list2.extend(list1)
print(list2)
```
