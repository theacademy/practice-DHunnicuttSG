```python
# While loop

i = 1
while i < 6:
  print(i)
  i += 1

# Do not forget to increment the loop

print('')  # blank line between outputs

# Break statement
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1

print('')  # blank line between outputs

# The continue Statement
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)

  # Why does 3 not print?

print('')  # blank line between outputs


  # Else statement

i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# For Loops

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)

# looping through a string
for x in "banana":
  print(x)

# Break statement
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
  
# Continue statement
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)

# The range function
# prints 0 to 5
for x in range(6):
  print(x)

# adding a start parameter 
# third parameter is step
for x in range(-4, 6):
  print(x)

for x in range(-4, 6, 2):
  print(x)

# else statement
# note:  else will NOT execute if the loop is stopped by a break statement
for x in range(6):
  print(x)
else:
  print("Finally finished!")

# nested loops
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```
